```python
def convert(num,to_base=10,from_base=10):
    if num=='':
        return 0
    num=str(num)
    if to_base==from_base:
        return str(num)
    if to_base==10:
        return str(int(num, base=from_base))
    if from_base==10:
        res=[]
        a='0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
        num=int(num)
        while num!=0:
            res.insert(0, a[num%to_base])
            num=num//to_base
        return ''.join(res)
    if from_base!=10 and to_base!=10:
        return convert(convert(num,from_base=from_base),to_base=to_base)
        
```

```python
def kaprekar_nat(n):
    square_n=n**2
    square_n=str(square_n)
    array_of_numerals=[]
    for numeral in square_n:
        array_of_numerals.append(numeral)
    len_of_array=len(array_of_numerals)
    first_term=''
    second_term=str(n**2)
    for some in array_of_numerals:
        first_term+=some
        second_term=second_term.lstrip(some)
        if second_term!='' and int(first_term)+int(second_term)==n:
            return True
    return False
```

```python
def kaprekar(num, base=10):
    if base==10:
        return kaprekar_nat(int(num))
    else:
        square_num=int(convert(num,from_base=base))**2
        square_num_base=convert(square_num,to_base=base)
        first_term=''
        second_term=square_num_base
        some=base
        for symbol in square_num_base:
            if int(convert(first_term, from_base=base))+int(convert(second_term, from_base=base))==int(num, base=some):
                return True
            first_term+=symbol
            second_term=''.join(second_term.split(symbol, 1))
            #second_term=second_term.lstrip[1:]
            if int(convert(first_term, from_base=base))+int(convert(second_term, from_base=base))==int(num, base=some):
                return True
        return False
```
